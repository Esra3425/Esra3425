```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas.api.types import CategoricalDtype
```

```python
sns.set(rc={'figure.figsize':(6,4)})
```

```python
df_nyc = pd.read_csv('AB_NYC_2019.csv')
df_nyc.head()
```

```python
df_nyc["neighbourhood_group"].value_counts().plot.barh().set_title("Frequencies of Neighbourhood Group")
```

```python
df_nyc["neighbourhood_group"].value_counts().plot.bar().set_title("Frequencies of Neighbourhood Group")
```

```python
df_nyc["neighbourhood_group"].index
```

```python
sns.barplot(x="room_type", y=df_nyc["room_type"].index, data=df_nyc, color="pink").set_title("Room Type Density")
```

```python
sns.barplot(x="room_type", y="price", hue="neighbourhood_group", data=df_nyc)
plt.title("room_type - price - neighbourhood_group")
```

```python
sns.barplot(x="price",
            y="room_type",
            hue="neighbourhood_group",
            data=df_nyc)
plt.title("price  - room_type - neighbourhood_group")
```

```python
df_nyc.info()
```

```python
df_nyc.groupby("room_type").sum().plot.pie(y="price", autopct="%.1f%%", ylabel="", legend=False, figsize=(5,5))
```

```python
cols = ["x", "y", "z"]

for col in cols:
 plt.figure(figsize=(5,5))
 sns.scatterplot(x="room_type", y="price", data=df_nyc)
```

```python
sns.histplot(df_nyc["price"], kde=True) # bins= auto
```

```python
sns.histplot(df_nyc["price"], kde=True, bins=40)
```

```python
sns.kdeplot(df_nyc["price"])
```

```python
sns.jointplot(x="price", y=df_nyc["room_type"], data=df_nyc)
```

```python
sns.catplot(x="room_type", y="price", data=df_nyc)
plt.title("room_type - Price")
```

```python
sns.catplot(x="room_type", y="price", hue="neighbourhood_group", kind="point", data=df_nyc)
plt.title("room_type - Price")
```

```python
sns.kdeplot(df_nyc["price"], shade=True)
```

```python
(sns.FacetGrid(df_nyc, hue="neighbourhood_group", height=5, xlim=(0, 10000)).map(sns.kdeplot, "price", shade=True).add_legend())
```
